```python
# Robust label handling, ensure numeric features, and safe xgboost fit
import numpy as np
import pandas as pd
import xgboost as xgb

# Ensure y are Series and aligned with X
y_train = pd.Series(y_train).reset_index(drop=True)
y_test = pd.Series(y_test).reset_index(drop=True)

# Drop rows with missing labels and keep X/Y aligned
mask_train = y_train.notnull()
mask_test  = y_test.notnull()
if not mask_train.all():
    drop_count = (~mask_train).sum()
    print(f'Dropping {drop_count} training rows with missing labels')
    X_train = X_train.loc[mask_train].reset_index(drop=True)
    y_train = y_train.loc[mask_train].reset_index(drop=True)
if not mask_test.all():
    drop_count = (~mask_test).sum()
    print(f'Dropping {drop_count} test rows with missing labels')
    X_test = X_test.loc[mask_test].reset_index(drop=True)
    y_test = y_test.loc[mask_test].reset_index(drop=True)

# Normalize label values to integers (0/1)
def normalize_labels(s):
    s = s.astype(str).str.strip()
    mapping = {'<=50K':'0','<=50k':'0','>50K':'1','>50k':'1','<=50 K':'0','>50 K':'1'}
    mapped = s.map(mapping).fillna(s)
    nums = pd.to_numeric(mapped, errors='coerce')
    if nums.isnull().any():
        bad = pd.Series(s)[nums.isnull()].unique()
        print('Could not convert these label values to numeric:', bad)
        raise ValueError('Non-convertible label values found')
    return nums.astype(int)

y_train = normalize_labels(y_train)
y_test  = normalize_labels(y_test)

print('Final unique classes in y_train:', np.unique(y_train))
print('Final unique classes in y_test :', np.unique(y_test))

# Convert any remaining object/categorical columns to numeric via one-hot encoding
X_train_enc = pd.get_dummies(X_train, dummy_na=False)
X_test_enc  = pd.get_dummies(X_test,  dummy_na=False)

# Align train/test columns (add missing columns with zeros)
X_train_enc, X_test_enc = X_train_enc.align(X_test_enc, join='outer', axis=1, fill_value=0)

# Ensure numeric dtypes (float) for xgboost
X_train = X_train_enc.astype(float).reset_index(drop=True)
X_test  = X_test_enc.astype(float).reset_index(drop=True)

print("X_train shape, X_test shape:", X_train.shape, X_test.shape)
print("Some dtypes (head):\n", X_train.dtypes.head())

# Re-create classifier (avoid stale attributes)
xgb_model = xgb.XGBClassifier(
    n_estimators=500,
    learning_rate=0.1,
    random_state=42,
    use_label_encoder=False,
    eval_metric='logloss'
)

# Try to build a callbacks list if available
callbacks = None
try:
    callbacks = [xgb.callback.EarlyStopping(rounds=10, save_best=True)]
except Exception:
    callbacks = None

# Fit model: try with callbacks first, fall back to fit without callbacks
if callbacks is not None:
    try:
        xgb_model.fit(X_train, y_train, eval_set=[(X_test, y_test)], callbacks=callbacks, verbose=True)
        print('Training completed successfully (with callbacks)')
    except TypeError as e:
        # older/newer xgboost may not accept callbacks param
        print('Callbacks not accepted by this XGBoost build:', e)
        print('Falling back to fit without callbacks...')
        xgb_model.fit(X_train, y_train, eval_set=[(X_test, y_test)], verbose=True)
        print('Training completed (without callbacks)')
else:
    xgb_model.fit(X_train, y_train, eval_set=[(X_test, y_test)], verbose=True)
    print('Training completed (no callbacks available)')
```

```python
# Make predictions on the training data
y_train_pred = xgb_model.predict(X_train)

# Make predictions on the test data
y_test_pred = xgb_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

# Calculate training accuracy
train_accuracy = accuracy_score(y_train, y_train_pred)

# Calculate testing accuracy
test_accuracy = accuracy_score(y_test, y_test_pred)

# Print the results
print(f"Training Accuracy: {train_accuracy:.4f}")
print(f"Testing Accuracy: {test_accuracy:.4f}")
```

## HYPERPARAMETER TUNING

```python
from sklearn.model_selection import GridSearchCV

```

```python
param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7]
}
print("Hyperparameter grid defined:")
print(param_grid)
```

```python
xgb_model_for_grid = xgb.XGBClassifier(random_state=42, eval_metric="logloss", use_label_encoder=False)

grid_search = GridSearchCV(
    estimator=xgb_model_for_grid,
    param_grid=param_grid,
    scoring='accuracy',
    cv=3,
    verbose=1,
    n_jobs=-1
)

print("GridSearchCV initialized with the XGBoost model and hyperparameter grid.")
```

```python
print("Fitting GridSearchCV to the training data...")
grid_search.fit(X_train, y_train)
print("GridSearchCV fitting complete.")
```

```python
print("Best parameters found: ", grid_search.best_params_)
print("Best accuracy score: ", grid_search.best_score_)
```

```python
import joblib

# Get the best parameters from grid_search
best_params = grid_search.best_params_

# Initialize a new XGBoost model with the best parameters
best_model = xgb.XGBClassifier(
    n_estimators=best_params['n_estimators'],
    learning_rate=best_params['learning_rate'],
    max_depth=best_params['max_depth'],
    random_state=42, # Ensure reproducibility
    eval_metric="logloss" # Use the same eval_metric as during tuning
)

# Train the best model on the full training data
print("Training the best XGBoost model...")
best_model.fit(X_train, y_train)
print("Best XGBoost model trained successfully.")

# Save the best model
model_filename = 'best_xgboost_model.joblib'
joblib.dump(best_model, model_filename)
print(f"Best XGBoost model saved as {model_filename}")
```
